# üìä Twitter Sentiment Analysis (50-row Sample Dataset)
This notebook loads the sentimentdataset.csv, calculates sentiment distribution, and plots bar, pie, and pico-style charts.

```python
# üì¶ Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# üìÇ Step 2: Load dataset (make sure it's in the same folder)
df = pd.read_csv("sentimentdataset.csv")
```

```python
# üë£ Step 2.5: Create df_50 from the original dataset
df_50 = df.head(50)  # first 50 rows
```

```python
# üëÅ Step 3: Preview and Basic Checks
print("üìÑ First 5 Rows of First 50 Entries:")
print(df_50.head())

print("\nüßæ Columns:", df_50.columns.tolist())
print("\nüßÆ Shape:", df_50.shape)
```

```python
# üßÆ Step 4: Sentiment Distribution
sentiment_counts = df['Sentiment'].value_counts()
print("\nüìä Sentiment Distribution:")
print(sentiment_counts)
```

```python
# üé® Step 5: Bar Chart - Smart Labeling for 50 rows

import matplotlib.pyplot as plt
import seaborn as sns

# Sort sentiments by frequency in the first 50 entries
order = df_50['Sentiment'].value_counts().index

# Set figure size
plt.figure(figsize=(10, 6))  # Slightly wider & taller for spacing

# Create bar chart
ax = sns.countplot(data=df_50, x='Sentiment', order=order, palette='Set2')

# Add better-positioned labels above bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f'{height}',                            # Label text
        (p.get_x() + p.get_width() / 2., height + 0.5),  # Position slightly above bar
        ha='center', va='bottom',
        fontsize=10, color='black'
    )

# Final touches
plt.title("Sentiment Count (First 50 Tweets)", fontsize=14)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Number of Tweets", fontsize=12)
plt.xticks(rotation=30, ha='right')  # Better label alignment
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```

```python
# üé® Step 6: Pie Chart - For First 50 Tweets

import matplotlib.pyplot as plt

# Count sentiment distribution for 50 rows
sentiment_counts_50 = df_50['Sentiment'].value_counts()

# Create pie chart
plt.figure(figsize=(6, 6))
colors = ['#66bb6a', '#ef5350', '#ffa726']  # Optional: Adjust for your 3 classes

sentiment_counts_50.plot.pie(
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    shadow=True,
    textprops={'fontsize': 12}
)

plt.title("Sentiment Distribution (First 50 Tweets)", fontsize=14)
plt.ylabel('')  # Remove y-axis label for cleaner look
plt.tight_layout()
plt.show()
```

```python
# üß± Step 7: Pico Chart (Text-based)
print("üß± Pico Chart:")
block = "‚ñ†"
for sentiment, count in sentiment_counts.items():
    blocks = block * (count // 10)  # Adjust for visual scale
    print(f"{sentiment:<10}: {blocks} ({count})")
```

```python
# üéØ Step 8: Bar Chart - First 50 Tweets

# Reuse sentiment_counts_50
plt.figure(figsize=(6, 4))
colors = ['#66bb6a', '#ef5350', '#ffa726']  # Same as pie chart for consistency

sentiment_counts_50.plot.bar(color=colors)

plt.title("Sentiment Count (First 50 Tweets)", fontsize=14)
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# üîÑ Step 9: Compare full vs. sample (Optional)

import matplotlib.pyplot as plt

# Define sentiment counts
sentiment_counts_100 = df.head(100)['Sentiment'].value_counts()
sentiment_counts_50 = df.head(50)['Sentiment'].value_counts()

# Colors
colors = ['#66bb6a', '#ef5350', '#ffa726']

# Plot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Pie - First 100
axs[0].pie(
    sentiment_counts_100,
    labels=sentiment_counts_100.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    shadow=True,
    textprops={'fontsize': 5}
)
axs[0].set_title("First 100 Tweets")

# Pie - Full Dataset
axs[1].pie(
    sentiment_counts_50,
    labels=sentiment_counts_50.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    shadow=True,
    textprops={'fontsize': 10}
)
axs[1].set_title("First 50 Tweets")

plt.tight_layout()
plt.show()


```
